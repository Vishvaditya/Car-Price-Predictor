```python
final_data1 = final_data.dropna()
print(final_data1.shape)
final_data.head()
```

```python
final_data1.isnull().sum()
```

```python
final_data = final_data1
```

```python
mileage = final_data['mileage']
mileage = mileage.tolist()
mileage
```

```python
mileage_num = []
for value in mileage:
    value_list = value.split()
    value_num = float(value_list[0])
    mileage_num.append(value_num)
print(mileage_num)
```

```python
mileage_num = pd.Series(mileage_num)
print(type(mileage_num))
```

```python
final_data['mileage_num'] = mileage_num
final_data.head()
```

```python
engine = final_data['engine']
engine = engine.tolist()
engine
```

```python
engine_num = []
for value in engine:
    value_list = value.split()
    value_num = float(value_list[0])
    engine_num.append(value_num)
print(engine_num)
```

```python
engine_num = pd.Series(engine_num)
final_data['engine_num'] = engine_num
final_data.head()
```

```python
max_power = final_data['max_power']
max_power = max_power.tolist()
max_power
```

```python
max_power_num = []
for value in max_power:
    value_list = value.split()
    value_num = float(value_list[0])
    max_power_num.append(value_num)
print(max_power_num)
```

```python
max_power_num = pd.Series(max_power_num)
final_data['max_power_num'] = max_power_num
final_data.head()
```

```python
final_data = final_data.drop(['mileage', 'engine', 'max_power', 'torque'], axis=1)
final_data.head()
```
