```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('car_sell_data.csv')
```

```python
data.head()
```

```python
print("Shape of dataframe : ", data.shape)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data['car_age'] = 2021 - data['year']
data.head()
```

```python
data.columns
```

```python
final_data = data[['selling_price', 'km_driven', 'fuel', 'seller_type',
       'transmission', 'owner', 'mileage', 'engine', 'max_power', 'torque',
       'seats', 'car_age']]
```

```python
print(final_data['fuel'].unique())
print(final_data['seller_type'].unique())
print(final_data['transmission'].unique())
print(final_data['owner'].unique())
```

```python
mileage = data['mileage']
mileage = mileage.tolist()
```

```python
print(final_data.shape)
print(final_data.isnull().sum())
```

```python
final_data1 = final_data.dropna()
print(final_data1.shape)
final_data.head()
```

```python
final_data1.isnull().sum()
```

```python
final_data = final_data1
```

```python
mileage = final_data['mileage']
mileage = mileage.tolist()
```

```python
mileage_num = []
for value in mileage:
    value_list = value.split()
    value_num = float(value_list[0])
    mileage_num.append(value_num)
```

```python
mileage_num = pd.Series(mileage_num)
```

```python
final_data['mileage_num'] = mileage_num
final_data.head()
```

```python
engine = final_data['engine']
engine = engine.tolist()
```

```python
engine_num = []
for value in engine:
    value_list = value.split()
    value_num = float(value_list[0])
    engine_num.append(value_num)
```

```python
engine_num = pd.Series(engine_num)
final_data['engine_num'] = engine_num
final_data.head()
```

```python
max_power = final_data['max_power']
max_power = max_power.tolist()
```

```python
max_power_num = []
for value in max_power:
    value_list = value.split()
    value_num = float(value_list[0])
    max_power_num.append(value_num)
```

```python
max_power_num = pd.Series(max_power_num)
final_data['max_power_num'] = max_power_num
final_data.head()
```

```python
final_data = final_data.drop(['mileage', 'engine', 'max_power', 'torque'], axis=1)
final_data.head()
```

```python
final_dataset = pd.get_dummies(final_data, drop_first=True)
final_dataset.head()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(final_dataset.corr(method='pearson'), cbar=False, annot=True, fmt='.1f', cmap='coolwarm')
```

```python
final_dataset = final_dataset.drop(['fuel_Petrol'], axis=1)
```

```python
final_dataset.head()
```

```python
final_dataset = final_dataset.drop(['max_power_num'], axis=1)
```

```python
final_dataset.isnull().sum()
```

```python
final_dataset = final_dataset.dropna()
```

```python
X = final_dataset.iloc[:,1:]
y = final_dataset.iloc[:, 0]
```

```python
X.head()
```

```python
y.head()
```

```python
X.isna().sum()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
model.feature_importances_
```

```python
feat_imp = pd.Series(model.feature_importances_, index=X.columns)
feat_imp.plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num=6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split,
              'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, scoring='neg_mean_squared_error', n_iter=10, cv=5, verbose=2, random_state=42, n_jobs=1)
```

```python
rf_random.fit(X_train, y_train)
```

```python
predictions = rf_random.predict(X_test)
predictions
```

```python
sns.distplot(y_test-predictions)
```

```python
import pickle
file = open("model.pkl", 'wb')
pickle.dump(rf_random, file)
```
